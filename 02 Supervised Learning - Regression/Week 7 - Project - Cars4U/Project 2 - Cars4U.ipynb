# 1. Project - Cars4U

## 1.1 Context
There is a huge demand for used cars in the Indian Market today. As sales of new cars have slowed down in the recent past, the pre-owned car market has continued to grow over the past years and is larger than the new car market now. Cars4U is a budding tech start-up that aims to find footholes in this market.

In 2018-19, while new car sales were recorded at 3.6 million units, around 4 million second-hand cars were bought and sold. There is a slowdown in new car sales and that could mean that the demand is shifting towards the pre-owned market. In fact, some car sellers replace their old cars with pre-owned cars instead of buying new ones. Unlike new cars, where price and supply are fairly deterministic and managed by OEMs (Original Equipment Manufacturer / except for dealership level discounts which come into play only in the last stage of the customer journey), used cars are very different beasts with huge uncertainty in both pricing and supply. Keeping this in mind, the pricing scheme of these used cars becomes important in order to grow in the market.

As a senior data scientist at Cars4U, you have to come up with a pricing model that can effectively predict the price of used cars and can help the business in devising profitable strategies using differential pricing. For example, if the business knows the market price, it will never sell anything below it.


## 1.2 Objective
To explore and visualize the dataset, build a linear regression model to predict the prices of used cars, and generate a set of insights and recommendations that will help the business.

 
## 1.3 Key Questions
1. What are the key factors influencing the price of used cars?
2. Is there a good predictive model for the price of used cars? What does the performance assessment look like for such a model?

## 1.4 Data Description
**used_cars_data.csv**  contains the different attributes of used cars sold in different locations. The detailed data dictionary is given below.

**Data Dictionary**

- `S.No.`: Serial number
- `Name`: Name of the car which includes brand name and model name
- `Location`: Location in which the car is being sold or is available for purchase (cities)
- `Year`: Manufacturing year of the car
- `Kilometers_driven`: The total kilometers driven in the car by the previous owner(s) in km
- `Fuel_Type`: The type of fuel used by the car (Petrol, Diesel, Electric, CNG, LPG)
- `Transmission`: The type of transmission used by the car (Automatic/Manual)
- `Owner`: Type of ownership
- `Mileage`: The standard mileage offered by the car company in kmpl or km/kg
- `Engine`: The displacement volume of the engine in CC
- `Power`: The maximum power of the engine in bhp
- `Seats`: The number of seats in the car
- `New_Price`: The price of a new car of the same model in INR Lakhs (1 Lakh INR = 100,000 INR)
- `Price`: The price of the used car in INR Lakhs

## 4.2 Data Manipulations

### 4.2.1 Process columns containing kmpl, CC and bhp

There are some columns that represent mileage, engine capacit and horsepower. The values all begin with numbers and end with the corresponding units like `kmpl`, `CC` and `bhp`. Since all the three columns have the same pattern, they can be processed together and converted into numbers. 

```python
def remove_suffix(column_val):
    """This function takes in a string representing a mileage, engine
    or power and converts it to a number. For example, '12.99 kmpl' 
    becomes 12.99. If the input is already NaN, this function just returns np.nan."""
    if isinstance(column_val, str):  # checks if `column_val` is a string or not
        spl = column_val.split(" ")  # splitting the value by space
        return float(spl[0])
    else:
        return np.nan


cols_with_suffix = ["Mileage", "Engine", "Power"]
for col in cols_with_suffix:
    df[col] = df[col].str.strip()  # remove the leading and trailing spaces
    df[col] = df[col].apply(
        remove_suffix
    )  # apply remove suffix method to all the values in the column

df.head()
```

### 4.2.2 Split Name into two columns

Split the name into make and model

```python
# Land Rover has a space in the make name which is different from other make names.
# Fixing it by using - instead of space
df["Name"] = df["Name"].str.replace("Land Rover", "Land-Rover")
df[df["Name"].str.startswith("Land")].head()
```

```python
# remove the leading and trailing spaces
# also convert the name into Title text (ISUZU was in upper and lower case)
split_df = df.Name.str.strip().str.title().str.split(" ", n=1, expand=True)

df.drop(["Name"], axis=1, inplace=True)

df["Make"] = split_df[0].str.strip()
df["Model"] = split_df[1]

del split_df  # don't need to do this but can keep things tidy

df.info()
```

```python
# Function to create barplots that indicate percentage for each category.


def perc_on_bar(data, z):
    """
    plot
    feature: categorical feature
    the function won't work if a column is passed in hue parameter
    """

    total = len(data[z])  # length of the column
    plt.figure(figsize=(15, 5))
    plt.xticks(rotation=45)
    ax = sns.countplot(data[z], palette="Paired")
    for p in ax.patches:
        percentage = "{:.1f}%".format(
            100 * p.get_height() / total
        )  # percentage of each class of the category
        x = p.get_x() + p.get_width() / 2 - 0.05  # width of the plot
        y = p.get_y() + p.get_height()  # hieght of the plot

        ax.annotate(percentage, (x, y), size=10)  # annotate the percantage
    plt.show()  # show the plot
```

## 6.4 Binning

**Based on the Make and the price of the cars, we can group them into Economy, MidSize and Luzury cars**

```python
def get_category_for_value(value):
    """This function takes in a value of the car and 
    returns the corresponding category."""
    if value < 8:
        return "Economy"
    elif value > 16:
        return "Luxury"
    else:
        return "MidSize"


# Find the mean price for each car make and calculate the category based on the mean price
make_mean = df.groupby("Make").Price.mean().sort_values(ascending=False)
category_dict = {"Economy": [], "MidSize": [], "Luxury": []}

for index, value in make_mean.items():
    cat = get_category_for_value(value)
    category_dict[cat].append(index)

category_dict
```

```python
def get_category_for_make(make):
    """This function takes in a make of the car and 
    returns the corresponding category."""
    for key, value in category_dict.items():
        if make in value:
            return key


df["make_bin"] = df["Make"].apply(get_category_for_make)
df.drop(["Make"], axis=1, inplace=True)
df.drop(["Model"], axis=1, inplace=True)
df.head(5)
```

## 7.6 Model performance check

* We will be using metric functions defined in sklearn for RMSE, MAE, and $R^2$.
* We will define a function to calculate adjusted $R^2$.
* We will create a function that will print out all the above metrics in one go.

```python
# Adjusted R^2
def adj_r2(ind_vars, targets, predictions):
    r2 = r2_score(targets, predictions)
    n = ind_vars.shape[0]
    k = ind_vars.shape[1]
    return 1 - ((1 - r2) * (n - 1) / (n - k - 1))


# Model performance check
def model_perf(model, inp, out):

    y_pred = model.predict(inp)
    y_act = out.values

    return pd.DataFrame(
        {
            "RMSE": np.sqrt(mean_squared_error(y_act, y_pred)),
            "MAE": mean_absolute_error(y_act, y_pred),
            "R^2": r2_score(y_act, y_pred),
            "Adjusted R^2": adj_r2(inp, y_act, y_pred),
        },
        index=[0],
    )
```

```python
# Checking model performance on train set
print("Training Performance\n")
print(model_perf(lin_reg_model, x_train, y_train))

# Checking model performance on test set
print("\n\nTest Performance\n")
print(model_perf(lin_reg_model, x_test, y_test))
```
